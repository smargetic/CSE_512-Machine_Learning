```python
#precision recall curve
def generate_precision_recall(y_true, y_pred):
  precision, recall, thresholds = sklearn.metrics.precision_recall_curve(y_true, y_pred)
  plt.plot(recall, precision)
  plt.title("Precision Recall Curve")
  plt.xlabel('Recall')
  plt.ylabel('Precision')
  plt.show()
```

```python
import random
#Part 2
def get_random(neg_data, num_random=10000):
  neg_return_data = []
  neg_return_indecies = []
  for i in range(0,num_random):
    rand_num = random.randint(0, num_random-1)
    if(rand_num not in neg_return_indecies):
      neg_return_data.append(neg_data[rand_num])
    else:
      i = i-1

  return np.array(neg_return_data)

#train data is dic of neg and positive values
def hard_negative_mining(train_data, train_output, test_data, test_output, iterations=10):
  objective_values = []
  ap_values = []

  #all pos data pts
  pos_train_indecies = [i for i, x in enumerate(train_output) if x == 1]
  pos_train_data = [train_data[i] for i in pos_train_indecies]
  
  #all neg data pts
  neg_train_indecies = [i for i, x in enumerate(train_output) if x == -1]
  neg_train_data = [train_data[i] for i in neg_train_indecies]
  
  #only get 10000 pts
  neg_train_ = get_random(neg_train_data)
  all_neg = neg_train_
  

  #conjoin the two sets
  data_set = np.append(np.array(pos_train_data), neg_train_, 0)

  #output
  data_output = np.array([1 for i in range(0,len(pos_train_data))] + [-1 for i in range(0,len(neg_train_))])

  #train first model
  clf = sklearn.svm.LinearSVC(verbose = 1).fit(data_set,data_output)
  test_pred = clf.predict(test_data)

  #objective values
  obj = (np.linalg.norm(np.array(clf.coef_))**2)/2.0 #divide by 2 square??
  objective_values.append(obj)

  #ap values
  ap_values.append(sklearn.metrics.average_precision_score(test_output, test_pred))

  #create svm model with randomly initialized data
  for i in range(0,iterations):
    #all non support vectors
    decision_function = clf.decision_function(neg_train_)
    #decision_function = clf.decision_function(all_neg)
    support_vector_indices = np.where(np.abs(decision_function) <= 1)
    #support_vector_indices = np.where(np.abs(decision_function) <= 1 + 1e-15)
    #support_vectors = neg_train_[support_vector_indices]
    support_vectors = all_neg[support_vector_indices]

    #since neg train count was never greater than 10000, can do this to find hard neg
    decision_function_all = clf.decision_function(all_neg)
    #hard_neg_indecies = np.where(decision_function_all > 3)
    hard_neg_indecies = np.where(decision_function_all > 3)
    print(len(hard_neg_indecies))
    hard_neg = all_neg[hard_neg_indecies]

    #create new data set
    neg_train_ = np.concatenate((support_vectors, hard_neg),axis=0)
  
    data_set = np.concatenate((pos_train_data,neg_train_),axis=0)
    data_output =  [1 for i in range(0,len(pos_train_data))] + [-1 for i in range(0,len(neg_train_))]

    #retrain model
    clf = sklearn.svm.LinearSVC(verbose = 0).fit(data_set,data_output)
    clf.fit(data_set, data_output)
    test_pred = clf.predict(test_data)

    #objective and ap values
    objective_values.append((np.linalg.norm(np.array(clf.coef_))**2)/2.0)
    ap_values.append(sklearn.metrics.average_precision_score(test_output, test_pred)) #test= val for part 2

#    objective_values.append([clf.coef_,clf.intercept_])
  print("Objective Values: " + str(objective_values))
  print("Average Precisions: " + str(ap_values))
  plot_("Objective Values", objective_values)
  plot_("Average Precisions", ap_values)
  return clf

clf = hard_negative_mining(train_data, train_output, val_data, val_output)
```
